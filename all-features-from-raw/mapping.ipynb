```python
import json
import pandas as pd
bundle_mapping = json.load(open("bundle_mapping_json.json"))
bundle_features = pd.read_csv("bundle_features_with_depth.csv",index_col=[0])
bundle_features["mapped_bundle_id"] = bundle_features["bundle_id"].apply(lambda x: bundle_mapping[str(x)])
bundle_features.to_csv("bundle_features_with_depth_mapped_id.csv")
```
